```python
count_falsePosition = 0 
Pprev = 0

def falsePosition_method(f, Xl, Xu, error):
    
    global Pprev
    global count_falsePosition 
        
    #check sign condition 
    if np.sign(f(Xl)) == np.sign(f(Xu)):
        raise Exception("The scalars Xl and Xu do not bound a root")
    
    #calculating midpoint 
    Pnew = (Xu*f(Xl)-Xl*f(Xu))/(f(Xl)-f(Xu))
    print ('calculated mid point: ',Pnew)
    
    count_falsePosition +=1

    #checking for error 
    Calerror = np.abs((Pnew-Pprev))
    print ('Error at ',count_falsePosition,'iteration: ', Calerror)
    if Calerror < error :
        num_iteration = count_falsePosition
        count_falsePosition = 0
        Pprev = 0 
        return num_iteration
    
    #checking sign condition and assinging the calculated Xm to prev to then be able to calculate error
    elif np.sign(f(Xl)) == np.sign(f(Pnew)): 
        Pprev = Pnew
        print ('New interval at ',count_falsePosition,'iteration:')
        print ('[',Pnew,',',Xu,']')
        return falsePosition_method(f,Pnew, Xu, error)
     #if b and midpoint same sign then recursive call the function with midpoint(replacing b) and a
    elif np.sign(f(Xu)) == np.sign(f(Pnew)):
        Pprev = Pnew
        print ('New interval at ',count_falsePosition,'iteration:')
        print ('[',Xl,',',Pnew,']')
        return falsePosition_method(f, Xl, Pnew, error)
    
```

```python
count_secantmethod = 0 

def secant_method(f, Ximinus1, Xi, error):
    global count_secantmethod    
    
    count_secantmethod +=1
    
    Xiplus1 = Xi - f(Xi)*((Xi-Ximinus1)/(f(Xi)-f(Ximinus1)))
    print ('calculated Xi+1 at',count_secantmethod,'iteration: ',Xiplus1)
    
    Calerror = np.abs(Xiplus1-Xi)
    print ('Error at ',count_secantmethod,'iteration: ', Calerror)
    
    
    if Calerror<error:
        num_iteration = count_secantmethod
        count_secantmethod = 0
        return num_iteration
    else: 
        print ('[',Xi,',',Xiplus1,']')
        return secant_method(f,Xi,Xiplus1,error)
```

```python
count_newton = 0 

def newton_method(f,f_dash,Ximinus1, error):
    global count_newton
    
    count_newton +=1
    
    Xiplus1 = Ximinus1 - (f(Ximinus1) / f_dash(Ximinus1))

    print ('calculated Xi+1 at',count_newton,'iteration: ',Xiplus1)
    
    Calerror = np.abs(Xiplus1-Ximinus1)
    print ('Error at ',count_newton,'iteration: ', Calerror)
    
        
    if Calerror<error:
        num_iteration = count_newton
        count_newton = 0
        return num_iteration
    
    else: 
        return newton_method(f,f_dash,Xiplus1,error)
    
```

```python
count_fixedpoint = 0 
Xprev = 0;

def fixedpoint_method(g,Ximinus1):
    global count_fixedpoint
    
    Xprev = Ximinus1
    
    count_fixedpoint +=1
    
    Xiplus1 = g(Ximinus1)
    
    print('X at iteration ',count_fixedpoint,'is: ',Xiplus1)
    
    if Xiplus1 == Xprev:
        iteration = count_fixedpoint
        count_fixedpoint=0
        return iteration
    else:
        return fixedpoint_method(g,Xiplus1)
        
```
